```python
def getTopVenues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        url='https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, RADIUS, LIMIT)
        results = requests.get(url).json()['response']['groups'][0]['items']
        venues_list.append([(name,
                             lat,
                             lng,
                             v['venue']['name'],
                             v['venue']['location']['lat'],
                             v['venue']['location']['lng'],
                             v['venue']['categories'][0]['name']) for v in results])
    popular_venues=pd.DataFrame([item for venue_list in venues_list for item in venue_list]) 
    popular_venues.colums=['Neighborhood','Neighborhood.Lat','Neighborhood.Lng','Venue','Venue.Lat','Venue.Lng','Venue.Category']
    return(popular_venues)
```

```python
Toronto_TopVenues=getTopVenues(names=Main_df['Neighbourhood'],latitudes=Main_df['Latitude'],longitudes=Main_df['Longitude'])
Toronto_TopVenues.shape
```

Use one-hot-encoding to analyze each neighborgood.

```python
Toronto_onehot=pd.get_dummies(Toronto_TopVenues[['Venue.Category']], prefix="", prefix_sep="")
Toronto_onehot.head()
```

```python
Toronto_onehot['Hood Name']=Toronto_TopVenues['Neighborhood']
Toronto_onehot.head(2)
# There is already a unique category named 'Neighborhood.'
```

```python
fixed_col=[Toronto_onehot.columns[-1]]+list(Toronto_onehot.columns[:-1])
Toronto_onehot=Toronto_onehot[fixed_col]
Toronto_onehot.head()
```

```python
Toronto_onehot.shape
```

```python
Toronto_grouped = Toronto_onehot.groupby('Hood Name').mean().reset_index()
Toronto_grouped
```

```python
Toronto_grouped.shape
```

```python
num_top_venues = 5

for each in Toronto_grouped['Hood Name']:
    print("----"+each+"----")
    temp = Toronto_grouped[Toronto_grouped['Hood Name'] == each].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

Put this into a pandas dataframe.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories=row.iloc[1:]
    row_categories_sorted=row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues=10
indicators=['st','nd','rd']

columns=['Hood Name']

for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1,indicators[ind]))
    except: 
        columns.append('{}th Most Common Venue'.format(ind+1))
```

```python
# Creat a new dataframe
hood_venues_sorted=pd.DataFrame(columns=columns)
hood_venues_sorted['Hood Name']=Toronto_grouped['Hood Name']

for ind in np.arange(Toronto_grouped.shape[0]): #39 rows in total
    hood_venues_sorted.iloc[ind,1:]=return_most_common_venues(Toronto_grouped.iloc[ind,:],num_top_venues)
```

```python
hood_venues_sorted.head(2)
```

Cluster Neighborhoods into 5 Clusters.

```python
from sklearn.cluster import KMeans
```

```python
kcluster=5
Toronto_grouped_clustering=Toronto_grouped.drop('Hood Name',axis=1)
kmeans=KMeans(n_clusters=kcluster, random_state=0).fit(Toronto_grouped_clustering)
kmeans.labels_[0:10]
```

```python
hood_venues_sorted.insert(0,'Cluster Labels',kmeans.labels_)
hood_venues_sorted.head(2)
```

```python
# In order to merge dataframes, I changed a column name in Main_df.
fix_columns=['Postal Code', 'Borough', 'Hood Name', 'Latitude', 'Longitude']
Main_df.columns=fix_columns
Main_df.head()
```

```python
Toronto_merged=Main_df
Toronto_merged=Toronto_merged.join(hood_venues_sorted.set_index('Hood Name'), on='Hood Name')
Toronto_merged.head()
```
